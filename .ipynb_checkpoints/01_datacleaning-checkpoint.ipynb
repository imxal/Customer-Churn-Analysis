The dataset contains Four columns that has missing values:Offer, Internet Type, Churn Category and Churn Reason.
Initial investigation suggests that some missing values are not data quality issues but rather represent business conditions. For example, churn-related fields are expected to be blank for customers who have not churned.

```python
df["Customer Status"].value_counts()
```

```python
df[df["Churn Category"].isnull()]["Customer Status"].value_counts()
```

```python
df[df["Churn Reason"].isnull()]["Customer Status"].value_counts()
```

#For Churn Reason and Churn Category         
Missing values in Churn Category and Churn Reason are expected because these variables are only populated for customers who have churned.

Therefore, the missing values represent a valid business condition rather than a data quality issue and were retained.

```python
#For Internet Type
df[df["Internet Type"].isnull()]["Internet Service"].value_counts()
```

```python
df["Internet Type"].value_counts()
```

Every missing value in the Internet Type may belong to customers who dont have internet service.

```python
df["Internet Type"]=(df["Internet Type"].fillna("No Internet"))
#filling missing values with value=no internet 
```

```python
df["Internet Type"].value_counts(dropna=False)
```

### Internet Type

The Internet Type variable contained 1,526 missing values.

Investigation revealed that all missing values corresponded to customers whose Internet Service status was "No".

Because these missing values represent a valid business condition rather than a data quality issue, they were replaced with the category "No Internet" to preserve business meaning and improve downstream analysis.

```python
df["Offer"].value_counts(dropna=False)
```

```python
df["Offer"]= df["Offer"].fillna("No Offer")
```

```python
df["Offer"].value_counts(dropna=False)
```

### Offer Variable Investigation

The Offer variable contained 3,877 missing values (55.05% of all records).

Based on the dataset documentation, the Offer variable represents the last marketing offer accepted by a customer, if applicable.

Therefore, missing values indicate that no marketing offer was accepted rather than a data quality issue.

To preserve business meaning, missing values were replaced with the category "No Offer".
