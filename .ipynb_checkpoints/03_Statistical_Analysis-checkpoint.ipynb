The purpose of this analysis is to determine whether key customer characteristics are statistically associated with customer churn.

The analysis will use hypothesis testing to evaluate differences between churned and retained customers.

```python
import pandas as pd
import numpy as np

from scipy import stats 
```

```python
df = pd.read_csv("data/processed/cleaned_telco.csv")
```

```python
df.head(10)
```

Do churned customers have significantly different monthly charges compared to retained customers?

## Null Hypothesis (H0): 
there is no significant difference in monthly charges between churned and retained customers.

## Alternative Hypothesis (H1):
there is a significant difference in monthly charges between churned and retained customers. 

```python
#creating groups
churned = df[df["Churn Label"] == "Yes"]["Monthly Charge"]
retained = df[df["Churn Label"] == "No"]["Monthly Charge"]
```

```python
#perform t-test
t_stat, p_value = stats.ttest_ind(
    churned,
    retained,
    equal_var= False )
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
#decision rule
alpha = 0.05
if p_value < alpha:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

Conclusion: there is a significant difference in monthly charges between churned and retained customers. 

Does customer satisfaction differ between churned and retained customers?

Null Hypothesis H0 :
There is no difference in satisfaction scores between churned and retained customers.

Alternative Hypothesis H1 :
There is a difference in satisfaction scores between churned and retained customers.

```python
#grouping
churned_satis = df[df["Churn Label"]== "Yes"]["Satisfaction Score"]
retained_satis = df[df["Churn Label"] == "No"]["Satisfaction Score"]
```

```python
#t-test
t_stat, p_value = stats.ttest_ind(
    churned_satis,
    retained_satis,
    equal_var=False)
print(p_value)
```

```python
alpha = 0.05
if p_value < alpha:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

Conclusion: There is a difference in satisfaction scores between churned and retained customers.

Offere Vs Customer Churn

Null Hypothesis H0:
Offer acceptance and customer churn are independent.There is no relationship between offer type and churn.

Alternative Hypothesis H1:
Offer acceptance and customer churn are associated.There is a relationship between offer type and churn.

```python
#Chi-Square Test

contingency = pd.crosstab(
    df["Offer"],
    df["Churn Label"]
)

chi2,p,dof,expected = stats.chi2_contingency(
    contingency
)
print(p)
```

```python
alpha = 0.05

if p < alpha:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

Conclusion: Offer acceptance and customer churn are associated.There is a relationship between offer type and churn.
